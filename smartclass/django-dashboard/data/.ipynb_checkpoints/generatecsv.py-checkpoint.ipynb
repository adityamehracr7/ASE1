```python
loaddata = pd.read_csv('arabic.csv')
loaddata.columns.values[0] = 'SID'
loaddata
```

```python
loaddata
```

```python
def createrequireddata(loaddata):
    new_df = pd.DataFrame(columns=['exam_type', 'SID', 'student_name','marks', 'attendence'])
    rowIndex = 0
    for i in range(1,4):
        for j in range(len(loaddata)): 
            new_df.loc[rowIndex, 'marks'] = loaddata.loc[j][i]
            new_df.loc[rowIndex, 'exam_type'] = loaddata.columns.values[i]
            new_df.loc[rowIndex, 'SID'] = loaddata.loc[j][0]
            new_df.loc[rowIndex, 'attendence'] = loaddata.loc[j][4]
            rowIndex+=1
    return new_df        
            
createrequireddata(loaddata)
```
